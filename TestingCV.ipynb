### Testing Image Basics

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../Notables/Ewer.webp")
plt.imshow(img)


```

```python
new_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
plt.imshow(cv.flip(new_img, 0))
plt.imshow(cv.flip(new_img, -1))
```

```python
cv.rectangle(new_img, (550, 250), (1380, 1750), (255, 0, 0), 3)
plt.imshow(new_img)
```
